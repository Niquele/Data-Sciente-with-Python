Aprendendo recursos do MatPlotLib
Utilizando dois pontos 'Causa da morte' e 'Porcentagem' para desenhar um chart que dê para visualizar o dataset. 

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Pricipais causas de morte no Brasil em 2021
causa = 'Acidentes','Homicídios','Suicídios','Covid-19','Câncer'
#Porcentangem a cada 100k entre faixa etária de 15-24
porcentagem = [10.940,4.437,4.287,1.378,1.306]
```

```python
plt.figure(figsize=(10,10))
explode = (0.05,0,0,0,0)
plt.pie(porcentagem,labels=causa,explode=explode,autopct='%1.1f%%',startangle=70)
plt.axis('equal')
plt.title('Causas de óbito a cada 100k hab. - Faixa Etária 15-24 anos - Brasil 2021')
```
